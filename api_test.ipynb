```python
import requests

import json

import random


# Replace this with your URL for it to work

ngrok_url = "https://598597b68e0d.ngrok-free.app/"

endpoint = "/predict"

url = f"{ngrok_url}{endpoint}"


# Example data with 30 features

# This example represents a hypothetical case for the breast cancer model.

# The data is randomly generated for demonstration purposes.

example_data = {

    "features": [

        17.99, 10.38, 122.8, 1001.0, 0.1184, 0.2776, 0.3001, 0.1471, 0.2419,

        0.07871, 1.095, 0.3563, 7.039, 175.7, 0.006399, 0.04904, 0.05373,

        0.01587, 0.03003, 0.006193, 25.38, 17.33, 184.6, 2019.0, 0.1622,

        0.6656, 0.7119, 0.2654, 0.4601, 0.1189

    ]

}


# Add some randomness to the data to simulate different inputs
random_data = [x + random.uniform(-1, 1) for x in example_data['features']]

example_data['features'] = random_data


# Print the data you're sending

print("Sending request with the following data:")

print(json.dumps(example_data, indent=2))


try:

    response = requests.post(url, json=example_data)

    response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)



    # Parse the JSON response

    prediction = response.json().get('prediction')

    print("\n--- API Response ---")

    print("Prediction:", prediction)


except requests.exceptions.RequestException as e:

    print(f"\nError: Could not connect to the API. Please ensure the ngrok server is running and the URL is correct.")

    print(f"Details:Â {e}")
```
